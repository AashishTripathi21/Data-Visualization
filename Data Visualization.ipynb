## Scatterplot using Seaborn

## Jitter Plot

```python
### See the plot between the columns city_mpg and highway_mpg by using catplot() function
sns.catplot(x="city_mpg", y="highway_mpg", data=df)
```

```python
sns.catplot(x="fuel_type", y="price", data=df)
```

```python
### We can see that the plot is scattered, so to handle that, we can set the jitter to false
sns.catplot(x="fuel_type" ,y="price", jitter = False,  data=df)
### Jitter is the deviation from the true value
```

```python
### Another Jitter plot using different parameters
sns.catplot(x="symboling", y="normalized_losses", data=df)
```

## Hue Plot

```python
### We can use the hue parameter for the plot
sns.catplot(x="fuel_type", y="price", hue = "number_of_doors", data=df)
```

```python
###  We can see that the points are overlapping each other, to eliminate this situation, we can set kind = “swarm”
sns.catplot(x="fuel_type", y="price", kind = "swarm", data=df)
### Swarm uses an algorithm that prevents the points from overlapping and adjusts the points along the categorical axis
```

```python
sns.catplot(x="body_style", y="aspiration", kind = "swarm", data=df)
```

```python
sns.catplot(x="aspiration", y="price", kind = "swarm", data=df)
```

```python
sns.catplot(x="fuel_type", y="price", hue = "aspiration", kind = "swarm", data=df)
```

## Boxplot using seaborn

```python
### Another kind of plot we can draw is a boxplot which shows three quartile values of the distribution along with the end values
sns.catplot(x="fuel_type", y="price", kind = "box", data=df)
```

```python
### When we use hue semantic with boxplot, it is leveled along the categorical axis so they don’t overlap
sns.catplot(x="fuel_type", y="price", hue = "aspiration", kind = "box", data=df)
```

## Violin Plot using seaborn

```python
### We can also represent the different variables differently by using violin plots
sns.catplot(x="fuel_type", y="price", hue = "aspiration", kind = "violin", data=df)
```

```python
### We can also split the violin when the hue semantic parameter has only two levels, which could also be helpful in saving space on the plot
sns.catplot(x="fuel_type", y="price", hue = "aspiration", kind = "violin", split = True, data=df)
```

## Boxplot using seaborn

```python
### Boxplot operates on the full dataset and obtains the mean value by default
sns.catplot(x="fuel_type", y="price", hue = "aspiration", kind = "bar", data=df)
```

## Pointplot using seaborn

```python
### Pointplot points out the estimate value and confidence interval. It connects data from the same hue category. This helps in identifying how the relationship is changing in a particular hue category
sns.catplot(x="fuel_type", y="price", hue = "aspiration", kind = "point", data=df)
```

## Swarm plot using seaborn

## Plotting Univariate Distributions
    

```python
### By default, distplot() function draws histogram and fits a Kernel Density Estimate
sns.distplot(df.price)
```

## Histogram using Seaborn 

```python
### We can also add a rugplot in it instead of using KDE (Kernel Density Estimate), which means at every observation, it will draw a small vertical stick
sns.distplot(df.price, kde=False, rug = True)
```

## Plotting Bivariate Distributions

```python
### Bivariate means joint, so to visualize it, we use jointplot() function of seaborn library
sns.jointplot(x="city_mpg", y="highway_mpg", data=df)
### By default, jointplot draws a scatter plot
```

## Hexplot using Seaborn

```python
### Hexplot is a bivariate analog of histogram as it shows the number of observations that falls within hexagonal bins
sns.jointplot(x=df.city_mpg, y=df.highway_mpg, kind="hex", data = df)
### This is a plot which works with large dataset very easily. 
### To draw a hexplot, we’ll set kind attribute to hex
```

## KDE Plot using Seaborn

```python
### This could be achieved by using jointplot() function and setting the attribute kind to KDE
sns.jointplot(x="city_mpg", y="highway_mpg", data=df, kind="kde")
```

## Ridge Plot using seaborn

```python
## Initialize the FacetGrid object
g = sns.FacetGrid(df, row="fuel_type", hue="fuel_type", aspect=5, height=3)

# # Draw the densities in a few steps
g.map(sns.kdeplot, "price", shade=True, alpha=1, lw=3.5, bw=.2)
g.map(sns.kdeplot, "price", color="w", lw=2, bw=.2)
g.map(plt.axhline, y=0, lw=2)

# # Define and use a simple function to label the plot in axes coordinates
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, color=color, ha="left", va="center", transform=ax.transAxes)
g.map(label, "price")

# Set the subplots to overlap
g.fig.subplots_adjust(hspace=-.25)

# # Remove axes details that don't play well with overlap
g.set_titles("")
g.set(yticks=[])
g.despine(bottom=True, left=True)
```
